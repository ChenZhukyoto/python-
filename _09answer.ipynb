```python
# 練習問題10

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

def sin(x):
    return math.sin(x)

def d0sin(n):
    if(n % 2 == 0):
        return 0
    return (-1)**(n // 2)

def taylorsin(n,x):
    return sum([d0sin(i)*(x**i)/math.factorial(i) for i in range(n)])

def taylorsingraph(n):
    x = np.linspace(-10, 10, 1000)
    plt.axis([-10,10,-2,2.0])
    y = [sin(z) for z in x]
    plt.plot(x, y)
    for i in range(4):
        y = [taylorsin(i, z) for z in x] 
        plt.plot(x, y)
    plt.show()

taylorsingraph(40)
```

```python
def logp1(x):
    return math.log(x+1)

def d0logp1(n):
    if(n == 0): 
        return 0
    return math.factorial(n-1)*((-1)**(n+1))

def taylor(fun,n,x):
    return sum([fun(i)*(x**i)/math.factorial(i) for i in range(n)])

funs = dict()
funs['sin'] = (sin, d0sin, -10, 10, -2, 2)
funs['logp1'] = (logp1, d0logp1, -0.9, 10, -2, 2)

def taylorgraph(fname,n):
    (f, df, min, max, ymin, ymax) = funs[fname]
    x = np.linspace(min, max, 1000)
    plt.axis([min,max,ymin,ymax])
    y = [f(z) for z in x]
    plt.plot(x, y)
    for i in range(n):
        y = [taylor(df, i, z) for z in x] 
        plt.plot(x, y, label=str(i))
    plt.legend(loc='lower right')
    plt.show()

taylorgraph("sin", 10)
taylorgraph("logp1", 10)
```

```python
# 練習問題30
%matplotlib inline
import matplotlib.pyplot as plt
import math

t = 0
x = 1
y = 0
delta = 0.1 # この値を変えてみよう
timeline = []
ix = []
iy = []
for n in range(100):
    ix.append(x)
    iy.append(y)
    timeline.append(t)
    x, y = x + y * delta, y - x * delta
    t = t + delta
plt.plot(timeline, ix)

y = [math.cos(t) for t in timeline]
plt.plot(timeline, y)

plt.show
```

```python
# 練習問題40
%matplotlib inline
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D

ax = plt.subplot(111, projection='3d')

p, r, b = 10, 28, 8/3
x, y, z, t = 1, 1, 1, 0 
delta = 0.01
timeline = []
ix, iy, iz = [], [], []

for n in range(4000):
    ix.append(x)
    iy.append(y)
    iz.append(z)
    timeline.append(t)
    x, y, z = x + (-p*x + p*y) * delta, \
              y + (-x*z + r*x -y) * delta, \
              z + (x*y - b*z) * delta
ax.plot(ix, iy, iz)

plt.show()
```
