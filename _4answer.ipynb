```python
# 練習問題10
def repeat(f, n=2, x=0.5):
    for _ in range(n):
        x = f(x)
    return x

def square (x):
    return x * x
print(repeat(square))
print(repeat(square, x=1.2))
print(f"{repeat(square, 7, 1.2):f}")
```

```python
#練習問題20
def repeatfun(f, n):
    return(lambda x: repeat(f, n, x))
```

```python
# 練習問題30
f = lambda x: 2.9*x*(1-x)
g = repeatfun(f, 2)
h = repeatfun(f, 3)

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 1, 100);
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.show()
```
